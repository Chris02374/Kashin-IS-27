ПЗ 6: Вариант 8
1. Дан целочисленный список размера 10. Вывести все содержащиеся в данном списе четные числа в порядке убывания их индексов, а также их количество K.

2. Дан список размера N. Найти количество участков, на которых его элементы монотонно возрастают.

3. Дан список размера N. Заменить каждый элемент списка на среднее арифметическое этого элемента и его соседей.



```python
import random

spisok = []
i = 0
while i < 10:
    spisok.append(random.randint(0, 100))
    i += 1

print("Исходный список:")
print(spisok)

k = 0
i = 9
print("\nЧетные числа в порядке убывания индексов:")
while i >= 0:
    if spisok[i] % 2 == 0:
        print(f"Индекс {i}: {spisok[i]}")
        k += 1
    i -= 1

print(f"\nКоличество четных чисел K: {k}")
```

```python
import random

n = input("Введите размер списка N: ")

while type(n) != int:
    try:
        n = int(n)
    except ValueError:
        print("Неправильно ввели!")
        n = input("Введите размер списка N: ")

spisok = []
i = 0
while i < n:
    spisok.append(random.randint(0, 100))
    i += 1

print("Исходный список:")
print(spisok)

count = 0
i = 1
length = 1

while i < n:
    if spisok[i] > spisok[i-1]:
        length += 1
    else:
        if length > 1:
            count += 1
        length = 1
    i += 1

if length > 1:
    count += 1

print(f"Количество участков монотонного возрастания: {count}")
```

```python
import random

n = input("Введите размер списка N: ")

while type(n) != int:
    try:
        n = int(n)
    except ValueError:
        print("Неправильно ввели!")
        n = input("Введите размер списка N: ")

spisok = []
i = 0
while i < n:
    spisok.append(random.randint(0, 100))
    i += 1

print("Исходный список:")
print(spisok)

noviy_spisok = []
i = 0
while i < n:
    if i == 0:
        srednee = (spisok[i] + spisok[i+1]) / 2
    elif i == n-1:
        srednee = (spisok[i-1] + spisok[i]) / 2
    else:
        srednee = (spisok[i-1] + spisok[i] + spisok[i+1]) / 3

    noviy_spisok.append(round(srednee, 2))
    i += 1

print("Новый список (среднее арифметическое с соседями):")
print(noviy_spisok)
```

Практическая работа №10
Вариант 8.
В магазинах имеются следующие товары. Магнит – молоко, соль, сахар. Пятерочка – мясо, молоко, сыр. Перекресток – молоко, творог, сыр, сахар. Определить:
1. в каких магазинах нельзя приобрести сыр.
2. в каких магазинах можно приобрести одновременно молоко и сахар.  
3. в каких магазинах можно приобрести соль.

```python
# Задание 1: Магазины, в которых нет сыра
magnit = {'молоко', 'соль', 'сахар'}
pyaterochka = {'мясо', 'молоко', 'сыр'}
perekrestok = {'молоко', 'творог', 'сыр', 'сахар'}
all_shops = {'Магнит', 'Пятерочка', 'Перекресток'}

shops_with_cheese = set()

if 'сыр' in magnit:
    shops_with_cheese.add('Магнит')
if 'сыр' in pyaterochka:
    shops_with_cheese.add('Пятерочка')
if 'сыр' in perekrestok:
    shops_with_cheese.add('Перекресток')

shops_without_cheese = all_shops - shops_with_cheese

print("Задание 1. Магазины, в которых НЕТ сыра:")

if shops_without_cheese:
    print(shops_without_cheese)
else:
    print("Сыр есть во всех магазинах.")
```

```python
# Задание 2: Магазины, в которых есть И молоко, И сахар
magnit = {'молоко', 'соль', 'сахар'}
pyaterochka = {'мясо', 'молоко', 'сыр'}
perekrestok = {'молоко', 'творог', 'сыр', 'сахар'}
required_products = {'молоко', 'сахар'}

shops_with_both = set()

if required_products.issubset(magnit):
    shops_with_both.add('Магнит')
if required_products.issubset(pyaterochka):
    shops_with_both.add('Пятерочка')
if required_products.issubset(perekrestok):
    shops_with_both.add('Перекресток')
print("\nЗадание 2. Магазины, где можно купить и молоко, и сахар:")
if shops_with_both:
    print(shops_with_both)
else:
    print("Нет магазина, где есть и молоко, и сахар одновременно.")
```

```python
# Задание 3: Магазины, в которых есть соль

magnit = {'молоко', 'соль', 'сахар'}
pyaterochka = {'мясо', 'молоко', 'сыр'}
perekrestok = {'молоко', 'творог', 'сыр', 'сахар'}

shops_with_salt = set()

if 'соль' in magnit:
    shops_with_salt.add('Магнит')
if 'соль' in pyaterochka:
    shops_with_salt.add('Пятерочка')
if 'соль' in perekrestok:
    shops_with_salt.add('Перекресток')
print("\nЗадание 3. Магазины, в которых можно купить соль:")
if shops_with_salt:
    print(shops_with_salt)
else:
    print("Соль не нашлась ни в одном магазине.")
```

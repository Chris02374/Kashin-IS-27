ПЗ 4: Вариант 8.

1. Даны два целых числа A и B (A < B). Найти сумму квадратов всех целых чисел от A до B включительно.

2. Начальный вклад в банке равен 1000 руб. Через каждый месяц размер вклада увеличивается на P процентов от имеющейся суммы (P — вещественное число, 0< P <25). По данному P определить, через сколько месяцев размер вклада превысит 1100 руб., и вывести найденное количество месяцев K (целое число) и итоговый размер вклада S (вещественное число).




```python
a = input("Введите число A: ")
b = input("Введите число B: ")

while type(a) != int:
    try:
        a = int(a)
    except ValueError:
        print("Неправильно ввели!")
        a = input("Введите число A: ")

while type(b) != int:
    try:
        b = int(b)
    except ValueError:
        print("Неправильно ввели!")
        b = input("Введите число B: ")

if a >= b:
    print("Ошибка: должно быть A < B")
else:
    summa = 0
    while a <= b:
        summa += a * a
        a += 1
    print(f"Сумма квадратов всех целых чисел от A до B: {summa}")
```

```python
p = input("Введите процент P (0 < P < 25): ")

while type(p) != float:
    try:
        p = float(p)
    except ValueError:
        print("Неправильно ввели!")
        p = input("Введите процент P (0 < P < 25): ")

if p <= 0 or p >= 25:
    print("Ошибка: P должно быть в диапазоне 0 < P < 25")
else:
    s = 1000.0
    k = 0

    while s <= 1100:
        s = s + s * (p / 100)
        k += 1

    print(f"Количество месяцев K: {k}")
    print(f"Итоговый размер вклада S: {s:.2f} руб.")
```

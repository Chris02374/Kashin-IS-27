8. Создайте класс "Человек" с атрибутами "имя", "возраст" и "пол". Напишите метод, который выводит информацию о человеке в формате "Имя: имя, Возраст: возраст, Пол: пол".

```
class Human:
    """Класс, представляющий человека."""

    def __init__(self, name: str, age: int, gender: str):
        """
        Инициализация объекта Human.

        Args:
            name: Имя человека
            age: Возраст человека
            gender: Пол человека
        """
        self.name = name
        self.age = age
        self.gender = gender

    def print_info(self) -> None:
        """Выводит информацию о человеке в требуемом формате."""
        print(f"Имя: {self.name}, Возраст: {self.age}, Пол: {self.gender}")
if __name__ == "__main__":
    print("=" * 50)
    print("ПРАКТИЧЕСКОЕ ЗАНЯТИЕ №16 - ВАРИАНТ 8")
    print("КЛАСС 'ЧЕЛОВЕК'")
    print("=" * 50)

    # Создание объектов класса Human
    person1 = Human("Алексей", 25, "Мужской")
    person2 = Human("Мария", 30, "Женский")
    person3 = Human("Дмитрий", 19, "Мужской")
    person4 = Human("Екатерина", 42, "Женский")

    # Вывод информации о каждом человеке
    print("\n--- Информация о людях ---")
    person1.print_info()
    person2.print_info()
    person3.print_info()
    person4.print_info()

    print("\n" + "=" * 50)
    print("✅ ТЕСТ ЗАВЕРШЕН УСПЕШНО")
    print("=" * 50)
```

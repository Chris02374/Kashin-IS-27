Дана строка 'апельсины 45 991 63 100 12 яблоки 13 47 26 0 16', отражающая продажи продукции по дням в кг. Преобразовать информацию из строки в словари, с использованием функции найти среднее значение продаж по каждому виду продукции, результаты вывести на экран.

```python
data_str = 'апельсины 45 991 63 100 12 яблоки 13 47 26 0 16'

data_list = data_str.split()
print("Разделенные элементы строки:", data_list)

sales = {}

i = 0

while i < len(data_list):

    if not data_list[i].isdigit():
        product_name = data_list[i]
        sales[product_name] = []
        i += 1


        while i < len(data_list) and data_list[i].isdigit():
            sales[product_name].append(int(data_list[i]))
            i += 1
    else:
        i += 1
print("\nСловарь с продажами:")
print(sales)
def calculate_average(numbers):
    """Функция вычисляет среднее значение списка чисел"""
    if len(numbers) == 0:
        return 0
    return sum(numbers) / len(numbers)

print("\nСредние продажи по каждому виду продукции:")

for product, quantities in sales.items():
    average = calculate_average(quantities)
    print(f"{product}: {quantities} -> среднее: {average:.2f} кг")
print("\nСловарь средних значений продаж:")
average_sales = {}
for product, quantities in sales.items():
    average_sales[product] = calculate_average(quantities)
print(average_sales)
```

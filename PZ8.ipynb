 2. Дана строка 'апельсины 45 991 63 100 12 яблоки 13 47 26 0 16', отражающая продажи продукции по дням в кг. Преобразовать информацию из строки в словари, с использованием функции найти среднее значение продаж по каждому виду продукции, результаты вывести на экран.

```python
sample_dict = {
    "name": "Kelly",
    "age": 25,
    "salary": 8000,
    "city": "New york"
}

print("Исходный словарь:", sample_dict)

name = sample_dict.get("name")
salary = sample_dict.get("salary")

print("Извлеченные значения:")
print(f"name: {name}")
print(f"salary: {salary}")
```

```python
sales_string = 'апельсины 45 991 63 100 12 яблоки 13 47 26 0 16'

print("Исходная строка:", sales_string)

items = sales_string.split()

sales_dict = {}
current_product = None

for item in items:
    if item.isalpha():
        current_product = item
        sales_dict[current_product] = []
    else:

        sales_dict[current_product].append(int(item))
print("Словарь с продажами:", sales_dict)

def calculate_average(sales_list):
    if len(sales_list) == 0:
        return 0
    return sum(sales_list) / len(sales_list)
print("Средние продажи по каждому виду продукции:")
for product, sales in sales_dict.items():
    avg = calculate_average(sales)
    print(f"{product}: {avg:.2f} кг")
```

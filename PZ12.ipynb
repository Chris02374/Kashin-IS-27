```
"""
Программа для подсчёта количества пар соседних элементов,
произведение которых кратно 3.
Используется списковое включение.
"""

import random


def count_pairs_divisible_by_3(sequence: list) -> int:
    """
    Подсчитывает количество пар соседних элементов,
    произведение которых делится на 3.

    :param sequence: список целых чисел
    :return: количество подходящих пар
    """
    # Используем списковое включение: создаём список из 1 для каждой подходящей пары,
    # затем суммируем элементы списка
    pairs_count = sum(
        1 for i in range(len(sequence) - 1)
        if (sequence[i] * sequence[i + 1]) % 3 == 0
    )
    return pairs_count


def main():
    """Главная функция программы."""
    # Генерируем случайную последовательность для демонстрации
    n = 10  # длина последовательности
    sequence = [random.randint(1, 20) for _ in range(n)]

    print("=" * 50)
    print("Задание 1. Подсчёт пар с произведением, кратным 3")
    print("=" * 50)
    print(f"Сгенерированная последовательность (n={n}):")
    print(sequence)

    # Вывод всех пар для наглядности
    print("\nВсе пары соседних элементов и их произведения:")
    for i in range(len(sequence) - 1):
        prod = sequence[i] * sequence[i + 1]
        is_divisible = "ДА" if prod % 3 == 0 else "НЕТ"
        print(f"Пара ({sequence[i]}, {sequence[i+1]}) -> произведение = {prod} -> кратно 3? {is_divisible}")

    # Подсчёт количества подходящих пар
    result = count_pairs_divisible_by_3(sequence)
    print(f"\nРезультат: количество пар = {result}")


if __name__ == "__main__":
    main()
```

```
"""
Программа-генератор, преобразующая все буквенные символы в заглавные.
Используется конструкция yield.
"""


def uppercase_generator(data):
    """
    Генератор, который преобразует буквенные символы в заглавные.

    :param data: строка или список символов
    :yield: символ в верхнем регистре (если это буква) или исходный символ
    """
    for char in data:
        if char.isalpha():  # Проверяем, является ли символ буквой
            yield char.upper()
        else:
            yield char


def main():
    """Главная функция программы."""
    print("=" * 50)
    print("Задание 2. Генератор для преобразования букв в заглавные")
    print("=" * 50)

    # Пример 1: обычная строка
    test_string = "Hello, World! 123 Python rules."
    print(f"\nИсходная строка: {test_string}")

    # Используем генератор
    gen = uppercase_generator(test_string)
    result_string = ''.join(gen)
    print(f"Преобразованная строка: {result_string}")

    # Пример 2: список символов
    test_list = ['a', 'b', 'C', 'd', '!', '@', 'z', 'X']
    print(f"\nИсходный список: {test_list}")

    gen2 = uppercase_generator(test_list)
    result_list = list(gen2)
    print(f"Преобразованный список: {result_list}")

    # Пример 3: демонстрация работы генератора с for (ленивые вычисления)
    print("\nПошаговая демонстрация работы генератора:")
    gen3 = uppercase_generator("Cat 24 Dog")
    for original, transformed in zip("Cat 24 Dog", gen3):
        print(f"'{original}' -> '{transformed}'")


if __name__ == "__main__":
    main()
```

Вариант 8.
 1. В двумерном списке элементы столбца N (N задать с клавиатуры) увеличить в два раза. 2. В двумерном списке элементы последней строки заменить на 0.

```
import random

# Исходная матрица (для примера 5x5)
rows, cols = 5, 5
matrix = [[random.randint(1, 9) for _ in range(cols)] for _ in range(rows)]

print("Исходная матрица:")
for row in matrix:
    print(row)

# Ввод номера столбца (нумерация с 0)
while True:
    try:
        N = int(input(f"\nВведите номер столбца для удвоения (0 до {cols-1}): "))
        if 0 <= N < cols:
            break
        else:
            print(f"Ошибка: введите число от 0 до {cols-1}")
    except ValueError:
        print("Ошибка: введите целое число")

# Удваиваем элементы столбца N (через списковое включение)
# Вариант: создаём новую матрицу
matrix_doubled = [row[:] for row in matrix]  # копия
for row in matrix_doubled:
    row[N] *= 2

# Альтернативный вариант (изменение исходной матрицы в цикле)
# for i in range(rows):
#     matrix[i][N] *= 2

print(f"\nРезультат (столбец {N} удвоен):")
for row in matrix_doubled:
    print(row)
```

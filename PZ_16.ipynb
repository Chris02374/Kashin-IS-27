Блок заданий 1:
8. Создайте класс "Человек" с атрибутами "имя", "возраст" и "пол". Напишите метод, который выводит информацию о человеке в формате "Имя: имя, Возраст: возраст, Пол: пол".

```python
class Chelovek:
    def __init__(self, imya, vozrast, pol):
        self.imya = imya
        self.vozrast = vozrast
        self.pol = pol

    def vyvesti_info(self):
        print(f"Имя: {self.imya}, Возраст: {self.vozrast}, Пол: {self.pol}")


# Создание объектов и проверка
chelovek1 = Chelovek("Иван", 22, "Мужской")
chelovek2 = Chelovek("Екатерина", 19, "Женский")
chelovek3 = Chelovek("Дмитрий", 35, "Мужской")

chelovek1.vyvesti_info()
chelovek2.vyvesti_info()
chelovek3.vyvesti_info()

chelovek1.vozrast = 23
chelovek1.vyvesti_info()
```

Блок заданий 2:
8. Создание базового класса "Животное" и его наследование для создания классов "Собака" и "Кошка". В классе "Животное" будут общие методы, такие как "дышать" и "питаться", а классы-наследники будут иметь свои уникальные методы и свойства, такие как "гавкать" и "мурлыкать".

```python
class Zhivotnoe:
    def __init__(self, imya, vid):
        self.imya = imya
        self.vid = vid
        self.energiya = 100

    def dyshat(self):
        print(f"{self.imya} ({self.vid}) делает вдох и выдох")
        return self

    def pitatsya(self, kolichestvo=10):
        self.energiya += kolichestvo
        print(f"{self.imya} поел, энергия теперь {self.energiya}")
        return self

    def spatsya(self):
        self.energiya += 20
        print(f"{self.imya} поспал, энергия теперь {self.energiya}")
        return self

    def info(self):
        print(f"Животное: {self.imya}, Вид: {self.vid}, Энергия: {self.energiya}")


class Sobaka(Zhivotnoe):
    def __init__(self, imya, poroda, vozrast):
        super().__init__(imya, "Собака")
        self.poroda = poroda
        self.vozrast = vozrast
        self.komandy = []

    def gavkat(self, gromko=True):
        if gromko:
            print(f"{self.imya}: ГАВ-ГАВ-ГАВ!!!")
        else:
            print(f"{self.imya}: гав-гав...")
        return self

    def vyuchit_komandu(self, komanda):
        self.komandy.append(komanda)
        print(f"{self.imya} выучил команду '{komanda}'")
        return self

    def vipolnit_komandu(self, komanda):
        if komanda in self.komandy:
            print(f"{self.imya} выполняет команду '{komanda}'")
            self.energiya -= 5
        else:
            print(f"{self.imya} не знает команду '{komanda}'")
        return self

    def info(self):
        komandy_str = ", ".join(self.komandy) if self.komandy else "не знает команд"
        print(f"Собака: {self.imya}, Порода: {self.poroda}, Возраст: {self.vozrast} лет")
        print(f"Энергия: {self.energiya}, Команды: {komandy_str}")


class Koshka(Zhivotnoe):
    def __init__(self, imya, okras, kogti=True):
        super().__init__(imya, "Кошка")
        self.okras = okras
        self.kogti = kogti
        self.nastroenie = "нейтральное"

    def murchat(self, intensivnost=1):
        zvuk = "мур" * intensivnost
        print(f"{self.imya} мурлыкает: {zvuk}...")
        self.nastroenie = "довольное"
        return self

    def tochit_kogti(self, predmet="диван"):
        if self.kogti:
            print(f"{self.imya} точит когти о {predmet}")
            self.energiya -= 3
        else:
            print(f"{self.imya} не может точить когти (когти подстрижены)")
        return self

    def igrat(self, igrushka="клубок"):
        print(f"{self.imya} играет с {igrushka}")
        self.energiya -= 5
        self.nastroenie = "игривое"
        return self

    def info(self):
        kogti_str = "есть" if self.kogti else "отсутствуют"
        print(f"Кошка: {self.imya}, Окрас: {self.okras}, Когти: {kogti_str}")
        print(f"Энергия: {self.energiya}, Настроение: {self.nastroenie}")

print("=" * 40)
print("ТЕСТИРОВАНИЕ КЛАССА СОБАКА")
print("=" * 40)

sobaka1 = Sobaka("Рекс", "Немецкая овчарка", 3)
sobaka2 = Sobaka("Тузик", "Дворняга", 1)

sobaka1.info()
sobaka1.dyshat().pitatsya(15).gavkat(True)
sobaka1.vyuchit_komandu("Сидеть")
sobaka1.vyuchit_komandu("Лежать")
sobaka1.vipolnit_komandu("Сидеть")
sobaka1.vipolnit_komandu("Голос")
sobaka1.spatsya()
sobaka1.info()

print("\n" + "=" * 40)
print("ТЕСТИРОВАНИЕ КЛАССА КОШКА")
print("=" * 40)

koshka1 = Koshka("Мурзик", "Черный", True)
koshka2 = Koshka("Снежок", "Белый", False)

koshka1.info()
koshka1.dyshat().pitatsya(8)
koshka1.murchat(3).igrat("мышь").tochit_kogti("кресло")
koshka1.spatsya()
koshka1.info()

print("\n" + "=" * 40)
print("ПРОВЕРКА РАБОТЫ С НЕСКОЛЬКИМИ ОБЪЕКТАМИ")
print("=" * 40)

zhivotnye = [sobaka1, sobaka2, koshka1, koshka2]
for zhivotnoe in zhivotnye:
    zhivotnoe.info()
    print("-" * 30)
```

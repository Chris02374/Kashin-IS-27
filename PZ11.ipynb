1. Средствами языка Python сформировать текстовый файл (.txt), содержащий последовательность из целых положительных и отрицательных чисел. Сформировать новый текстовый файл (.txt) следующего вида, предварительно выполнив требуемую обработку элементов:

Исходные данные: Количество элементов: Индекс последнего минимального элемента: Сумма элементов больших 10 во второй половине:  

2. Из предложенного текстового файла (text18-8.txt) вывести на экран его содержимое, количество символов, принадлежащих к группе букв. Сформировать новый файл, в который поместить текст в стихотворной форме предварительно удалив букву «с» из текста.


```python
import random

# 1. Создаём исходный файл с числами
numbers = [random.randint(-50, 50) for _ in range(15)]  # 15 случайных чисел
with open('numbers_8.txt', 'w', encoding='utf-8') as f:
    f.write(' '.join(map(str, numbers)))

print("Исходный файл numbers_8.txt создан.")

# 2. Чтение и обработка
with open('numbers_8.txt', 'r', encoding='utf-8') as f:
    data = f.read().strip().split()
    data = list(map(int, data))

n = len(data)
print("Исходные данные:", data)
print("Количество элементов:", n)

# Поиск последнего минимального элемента
min_value = min(data)
last_min_index = max(i for i, val in enumerate(data) if val == min_value)

# Вторая половина (если нечётное количество — вторая половина начинается с n//2)
half_start = n // 2
second_half = data[half_start:]
sum_gt_10 = sum(x for x in second_half if x > 10)

# 3. Запись результатов в новый файл
with open('result_8.txt', 'w', encoding='utf-8') as f:
    f.write("Исходные данные: " + ' '.join(map(str, data)) + '\n')
    f.write("Количество элементов: " + str(n) + '\n')
    f.write("Индекс последнего минимального элемента: " + str(last_min_index) + '\n')
    f.write("Сумма элементов больших 10 во второй половине: " + str(sum_gt_10))

print("Результат сохранён в файл result_8.txt")
print(f"Индекс последнего минимального: {last_min_index}")
print(f"Сумма >10 во второй половине: {sum_gt_10}")
```

```python
# Так как в Colab нет исходного файла, создаём его из вашего текста
poem_text = """И только небо засветилось,
Все шумно вдруг зашевелилось,
Сверкнул за строем строй.
Полковник наш рожден был хватом:
Слуга царю, отец солдатам…
Да, жаль его: сражен булатом,
Он спит в земле сырой."""

# Сохраняем в файл
with open('text18-8.txt', 'w', encoding='utf-8') as f:
    f.write(poem_text)

print("=== Содержимое файла text18-8.txt ===\n")
with open('text18-8.txt', 'r', encoding='utf-8') as f:
    lines = f.readlines()
    for line in lines:
        print(line, end='')

# Считаем количество символов-букв
letter_count = 0
with open('text18-8.txt', 'r', encoding='utf-8') as f:
    content = f.read()
    for ch in content:
        if ch.isalpha():  # буквы любых алфавитов
            letter_count += 1

print("\nКоличество символов, принадлежащих к группе букв:", letter_count)

# Удаляем буквы 'с' и 'С', результат в новый файл
content_no_c = content.replace('с', '').replace('С', '')

with open('text18-8_no_c.txt', 'w', encoding='utf-8') as f:
    f.write(content_no_c)

print("Файл text18-8_no_c.txt создан (буква 'с' удалена).\n")
print("=== Содержимое нового файла ===\n")
print(content_no_c)
```

ПЗ 3: Вариант 8.
 1. Даны два целых числа: A, B. Проверить истинность высказывания: «Каждое из чисел A и B нечетное».

 2. Единицы длины пронумерованы следующим образом: 1 — дециметр, 2 — километр, 3 — метр, 4 — миллиметр, 5 — сантиметр. Дан номер единицы длины (целое число в диапазоне 1-5) и длина отрезка в этих единицах (вещественное число). Найти длину отрезка в метрах

```python
a = input("Введите число A: ")
b = input("Введите число B: ")

while type(a) != int:
    try:
        a = int(a)
    except ValueError:
        print("Неправильно ввели!")
        a = input("Введите число A: ")

while type(b) != int:
    try:
        b = int(b)
    except ValueError:
        print("Неправильно ввели!")
        b = input("Введите число B: ")

if a % 2 != 0 and b % 2 != 0:
    print("Высказывание «Каждое из чисел A и B нечетное» да")
else:
    print("Высказывание «Каждое из чисел A и B нечетное» нет")
```

```python
nomer = input("Введите номер единицы длины (1-5): ")
dlina = input("Введите длину отрезка: ")

while type(nomer) != int:
    try:
        nomer = int(nomer)
    except ValueError:
        print("Неправильно ввели номер!")
        nomer = input("Введите номер единицы длины (1-5): ")

while type(dlina) != float:
    try:
        dlina = float(dlina)
    except ValueError:
        print("Неправильно ввели длину!")
        dlina = input("Введите длину отрезка: ")

if nomer == 1:
    result = dlina * 0.1
elif nomer == 2:
    result = dlina * 1000
elif nomer == 3:
    result = dlina
elif nomer == 4:
    result = dlina * 0.001
elif nomer == 5:
    result = dlina * 0.01
else:
    print("Ошибка: номер должен быть от 1 до 5")
    exit()

print(f"Длина отрезка в метрах: {result}")
```
